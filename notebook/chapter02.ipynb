```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 500)
y1 = x**2 + 1
y2 = x
x3 = [x for x in x if x > 1]
y3 = [np.sqrt(x - 1) for x in x if x > 1]

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x3,y3)
plt.ylim(-2,4)
plt.grid()
plt.show()
```

```python
delta = 0.2
X = np.linspace(0.5 - delta, 0.5 + delta, 100)
Y = X**3 - X

plt.plot(X,Y)
plt.grid()
plt.show()
```
