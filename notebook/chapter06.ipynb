```python
%matplotlib inline
import numpy as np
import scipy.special as scm
import matplotlib.pyplot as plt

def gauss(x, n):
    m = n/2
    return np.exp(-(x - m)**2 / m) / np.sqrt(m * np.pi)

N = 1000
M = 2**N

X = range(440, 561)
plt.bar(X, [scm.comb(N, i) / M for i in X])
plt.plot(X, gauss(np.array(X), N), c='k', linewidth=2)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gauss_func(x,sigma):
    return np.exp(-x**2 / 2 /sigma**2) / np.sqrt(2 * np.pi) / sigma

def sigmoid_func(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-4,4,100)
plt.plot(x, gauss_func(x,1.6), c='k', linewidth=1)
plt.plot(x, sigmoid_func(x)*(1 -sigmoid_func(x)) , c='r', linewidth=2)
plt.show()
```
