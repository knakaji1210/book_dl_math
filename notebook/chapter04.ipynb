```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

z = 3*X**2 + 3*Y**2 - X*Y + 7*X - 7*Y +10

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X,Y,z)
ax.contour(X, Y, z, colors='black', offset=-1)
plt.show()
```

```python
# numpy.meshgridの勉強　（https://python.atelierkobato.com/meshgrid/）

x = np.arange(3)
print(x)
y = np.arange(3,6)
print(y)

# 3×3の格子点を作成
X, Y = np.meshgrid(x, y)
print("配列X\n{}\n".format(X))
print("配列X\n{}\n".format(Y))
```

```python
# 配列の要素数
n = 9

x = np.linspace(0, 5, n)
y = np.linspace(0, 5, n)

print(x)

# 5×5の格子点を作成
X, Y = np.meshgrid(x, y)

print("配列X\n{}\n".format(X))
print("配列Y\n{}".format(Y))
```

```python
# 2変数関数を定義
def f(x, y):
    return x**2 - y**2

# z=f(x,y)を計算
Z = f(X, Y)
Z_round = np.round(Z, 2)

print(Z_round)
```

```python
# Figureと3DAxesを追加
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection="3d")

# 3DAxesの設定
ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)
ax.view_init(45, 45)

# 曲面を描画
ax.plot_surface(X, Y, Z, cmap = "autumn")

plt.show()
```
