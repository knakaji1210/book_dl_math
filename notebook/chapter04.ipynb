```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

z = 3*X**2 + 3*Y**2 - X*Y + 7*X - 7*Y +10

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X,Y,z)
ax.contour(X, Y, z, colors='black', offset=-1)
plt.show()
```
