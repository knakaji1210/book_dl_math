```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 10)
y = np.power(2, x)

x2 = np.linspace(-2, 2, 100)
y2 = np.power(2, x2)

plt.scatter(x,y, color='black')
plt.plot(x2,y2)
plt.grid()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 500)
y1 = np.power(2, x)
y2 = x
x3 = [x for x in x if x > 0]
y3 = [np.log2(x) for x in x if x > 0]

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x3,y3)
plt.ylim(-2,4)
plt.grid()
plt.show()
```

```python
import numpy as np
np.set_printoptions(precision=10)
x = np.logspace(0, 11, 12, base=0.1, dtype='float64')
y = np.power(1+x, 1/x)
for i in range(11):
    print('x = %12.10f y = %12.10f' % (x[i],y[i]))
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.2, 2, 100)
y = x - 1
y1 = np.log2(x)
y2 = np.log(x)
y3 = np.log10(x)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.ylim(-1,1)
plt.grid()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6, 500)
y = 1/(1+np.exp(-x))

plt.plot(x,y)
plt.ylim(0,1)
plt.grid()
plt.show()
```
