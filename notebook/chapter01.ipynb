```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = [166, 176, 171, 173, 169]
y = [58.7, 75.7, 62.1, 70.4, 60.1]

plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
X = [x[i] - np.mean(x) for i in range(len(x))]
Y = [y[i] - np.mean(y) for i in range(len(y))]

plt.scatter(X,Y)
plt.grid()
plt.show()
```

```python
Xp = np.linspace(np.min(X),np.max(X),20)
Yp = Xp * 1.82068

plt.scatter(X,Y)
plt.plot(Xp,Yp, color='green')
plt.grid()
plt.show()
```
